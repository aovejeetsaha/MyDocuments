```python
class Patient:
  def __init__(self, name,id, age, blood_group):
    self.name = name
    self.id = id
    self.age = age
    self.blood_group = blood_group
    self.prev = None
    self.next = None

class WRM:
  def __init__(self):
    self.dh =  Patient(None,None,None,None)
    self.dh.next = self.dh
    self.dh.prev = self.dh

  def register_patient(self, name, id, age, blood_group):
    n_p = Patient(name, id, age, blood_group)

    if self.dh.next == None:
        self.dh.next = n_p
        self.dh.prev = n_p
        n_p.next = self.dh
        n_p.prev = self.dh
    else:
        temp = self.dh.prev
        temp.next = n_p
        n_p.next = self.dh
        n_p.prev = temp
        self.dh.prev = n_p

    print("---->Patient Added.")

  def serve_patient(self):

    if self.dh.next == self.dh:
      print("----> No patients in the waiting room.")

    else:

      temp = self.dh.next
      print("----> Serving patient:", temp.name)
      temp.prev.next = temp.next
      temp.next.prev = temp.prev


  def cancel_all(self):

    self.dh.next = self.dh
    self.dh.prev = self.dh

    print("----> All appointments are been canceled .")

  def can_doctor_go_home(self):
    flag = False

    if self.dh.next == self.dh:
      flag = True

    else:
      flag = False

    return flag

  def show_all_patients(self):

    if self.dh.next == self.dh:
      print("---> No patients in the waiting room.")

    else:
      p = self.dh.next

      while p != self.dh:
        print("--> Name:", p.name, "--> ID:", p.id)
        p = p.next



  def display_menu(self):
    print(f'1. Add Patient ')
    print(f'2. Serve Patient ')
    print(f'3. Show All Patient ')
    print(f'4. Can Doctor Go Home ? ')
    print(f'5. Cancle All Appiontments ')
    print(f'6. Exit ')


# Driver code

wrm = WRM()

wrm.display_menu()


while True:
  num = input("Enter your choice: ")

  if num == '1':
    name = input("Enter patient name: ")
    id = int(input("Enter patient ID: "))
    age = int(input("Enter patient age: "))
    blood_group = input("Enter patient blood group: ")

    wrm.register_patient(name, id, age, blood_group)

  elif num == '2':
    wrm.serve_patient()

  elif num == '3':
    wrm.show_all_patients()

  elif num == '4':
      if wrm.can_doctor_go_home():
        print("Yes, doctor can go home.")
      else:
         print("No, patients are still waiting.")

  elif num == '5':
    wrm.cancel_all()

  elif num == '6':
    print('Exiting the program')
    break

  else:
    print("Invalid number. Please try again.")
```
